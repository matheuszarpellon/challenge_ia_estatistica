```python
# Importar libs
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dicionario de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
# Montar Dataframe
df_faturamento = pd.DataFrame(dict_faturamento)
```

# Objetivos

* Trazer a média das vendas
* Criar um gráfico de barras vertical mostrando o mês de referência e o valor
* Criar um gráfico de linhas mostrando o mês de referência e o valor.

```python
# Visualizar Dataframe
df_faturamento
```

```python
# Trazer a média das vendas
media_vendas = round(df_faturamento['valor'].mean(), 2)
media_vendas
```

```python
# Criar um gráfico de barras vertical mostrando o mês de referência e o valor
df_faturamento.plot.bar(x='data_ref', y='valor', color='g', title='Vendas por mês', xlabel='Mês de referência', ylabel='Valor')
```

```python
# Criar um gráfico de linhas mostrando o mês de referência e o valor
df_faturamento.plot.line(x='data_ref', y='valor', color='g', title='Vendas por mês', xlabel='Mês de referência', ylabel='Valor')
```
